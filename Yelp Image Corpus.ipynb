```python
import pprint
import time

import numpy as np 
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

from scipy.misc import imread
from subprocess import check_output
from PIL import Image
from numpy import genfromtxt, savetxt


# from sklearn.neural_network import MLPClassifier

start_time = round(time.time())
tag = str(start_time)
%matplotlib inline
```

```python
train_photos = pd.read_csv('train_photo_to_biz_ids.csv')
train_labels = pd.read_csv('train.csv')
test_photos = pd.read_csv('test_photo_to_biz.csv')
```

```python
def img_as_array(image_id, test_indicator=False):
    if not test_indicator:
        imagepath = 'train_photos/' + str(image_id) + '.jpg'
    else:
        imagepath = 'test_photos/' + str(image_id) + '.jpg'
        
    img = imread(imagepath)
    img.resize(100, 100)
    return img
```

```python
plt.rcParams['figure.figsize'] = (10.0, 10.0)
plt.subplots_adjust(wspace=0, hspace=0)

for i in range(25):
    plt.subplot(5, 5, i + 1)
    img = Image.open('train_photos/' + str(train_photos.photo_id[i]) + '.jpg')
    img = img.resize((100, 100), Image.ANTIALIAS)
    plt.imshow(img)
    plt.axis('off')
```
