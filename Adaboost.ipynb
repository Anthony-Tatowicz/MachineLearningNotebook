## Ensemble Methods

For this notebook I will be going over an ensemble method known as adaboost. There are other ensemble methods like random forest, however the ensemble methods more or less use the same idea. 

### Basic Idea

Ensemble methods take advantage of using weak learners, say $t_1, t_2, ..., t_n$ and combines them to create a better model. The constraint is that each model $w$ has to be at least slightly better than random, otherwise the classifier would not be able to predict an outcome, since our entropy would be a maximum. $H(x) = 1$

### Example:

An example of this is taking multiple small decision trees and combining them to create a forest, that is what random forest does. 

### Adaboost

Adaboost is a method very similar to randomforest, in that it takes a sum of weak learners. The idea is to train weak learners by iterating over the data multiple times, but for each iteration the data is modified with a weight $w_1, w_2, ..., w_n$ for each of the training samples. The weights will change at each iteration based on the performance of the learner, so if one sample is repeatedly wrong the weight $w_i$ will increase.

### Classification on iris dataset

```python
from sklearn.cross_validation import cross_val_score
from sklearn.datasets import load_iris
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt

iris = load_iris()

# Take 90% for training 10% for testing
training_range = round(len(iris.data) * .9)

X_train = iris.data[:training_range]
y_train = iris.target[:training_range]
X_test = iris.data[training_range:]
y_test = iris.target[training_range:]

# arg of n_estimators=100, this is the number of weak learners
clf = AdaBoostClassifier(n_estimators=100)
clf.fit(X_train, y_train)

pred = clf.predict(X_test)

scores = cross_val_score(clf, iris.data, iris.target)
print('Cross Val: {:.2%}'.format(scores.mean()))
print('Acc Score: {:.2%}'.format(accuracy_score(pred, y_test)))
```

### Scoring

The Cross Validation is a better representation of the classifiers performance, because it shuffles and randomly splits the data, as where the accuracy score I split the data by simply taking the first 90% for training and the last 10% for testing.

### Plot

```python
%matplotlib inline
n_classes = 3
plot_colors = "bry"
plot_step = 0.02

for pairidx, pair in enumerate([[0, 1], [0, 2], [0, 3],
                                [1, 2], [1, 3], [2, 3]]):
    
    # We only take the two corresponding features
    X = iris.data[:, pair]
    y = iris.target

    # Shuffle
    idx = np.arange(X.shape[0])
    np.random.seed(13)
    np.random.shuffle(idx)
    X = X[idx]
    y = y[idx]

    # Standardize
    mean = X.mean(axis=0)
    std = X.std(axis=0)
    X = (X - mean) / std

    # Train
    clf = AdaBoostClassifier(n_estimators=100).fit(X, y)

    # Plot the decision boundary
    plt.subplot(2, 3, pairidx + 1)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)

    plt.xlabel(iris.feature_names[pair[0]])
    plt.ylabel(iris.feature_names[pair[1]])
    plt.axis("tight")

    # Plot the training points
    for i, color in zip(range(n_classes), plot_colors):
        idx = np.where(y == i)
        plt.scatter(X[idx, 0], X[idx, 1], c=color, label=iris.target_names[i],
                    cmap=plt.cm.Paired)

    plt.axis("tight")

plt.suptitle("Decision surface of a decision tree using paired features")
plt.show()
```

### Conclusion

As we can see Adaboost does well, we have a cross validation score of 96% and accuracy score of 100%, it also helps with the over fitting problem we faced with decision trees.
